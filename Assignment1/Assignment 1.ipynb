# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import math
```

```python
import mpld3
mpld3.enable_notebook()
```

# Taking input

```python
n =  np.arange(1,8,1)
k = 4
```

# Processing the array
### logarithm 
### linear 
### quardic
### cubic 
### polynominal 

```python
logarithm = []
linear = []
quardic = []
cubic = []
polynominal = []

for x in n:
    logarithm.append(math.log2(x))
    linear.append(x) 
    quardic.append(math.pow(x, 2))
    cubic.append(math.pow(x, 3))
    polynominal.append(math.pow(x,k))
    
```

```python

#figure(figsize=(8, 6), dpi=80)
figure = plt.figure(figsize=(10, 8), dpi=80)
# plot the linear data and the exponential data
plt.plot(n,logarithm, '-o',alpha=0.8,color = 'magenta')
plt.plot(n,linear, '-o',alpha=0.8, color = 'black')
plt.plot(n,quardic, '-o',alpha=0.8)
plt.plot(n,cubic, '-o',alpha=0.8)
plt.plot(n,polynominal, '-o',alpha=0.8)

plt.xlabel('Value of N')
plt.ylabel('Values of Y')
plt.title('Visualizing Data')
plt.legend(['Logarithm', 'Linear', 'Quardic', 'Cubic', 'Polynominal'])
plt.savefig("J:/figure.jpg")
```
